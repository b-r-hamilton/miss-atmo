## OPENDaP process

Get relevant metadata for chosen variables - here we look at air temperature at the 2m level, snow coverage, volumetric soil moisture, and water runoff

```python
from pydap.client import open_url 
import pandas as pd
url_base = 'http://psl.noaa.gov/thredds/dodsC/Datasets/20thC_ReanV3/Monthlies/'
var = ['2mMO/air.2m.mon.mean.nc', 'sfcMO/csnow.mon.mean.nc', 'subsfcMO/soilw.mon.mean.nc', 'accumsMO/runoff.mon.mean.nc', 'sfcMO/pres.sfc.mon.mean.nc']

not_var = ['lat', 'lon', 'time', 'time_bnds', 'level']
var_df = pd.DataFrame(index = ['description', 'location', 'units', 'level', 'missing_val'])
for v in var: 
    url = url_base + v
    x = open_url(url)
    keys = list(x.keys())
    level_bool = False
    if 'level' in keys: level_bool = True
    varname = [k for k in keys if k not in not_var][0]
    desc = x[varname].attributes['var_desc']
    na = x[varname].attributes['missing_value']
    var_df[varname] = [desc, v, x[varname].units, level_bool, na]
    
var_df
```

Get parameter values - lat, lon, and level, if it exists 

```python
#are there any variables with 'level' data? What rows?
if True in list(var_df.loc['level']):
    index = list(var_df.loc['level']).index(True)
varname = var_df.iloc[:, index].name #find variable name that has a True for 'level'
url = url_base + var_df[varname]['location']
x = open_url(url)

par_df = {}
for v in not_var:
    par_df[v] = x[v][:].data
```

```python
import datetime as dt 
origin_string = x['time'].attributes['units']
time = [em.convert_datetime(t, origin_string) for t in par_df['time']]
par_df['time'] = time
```

```python
riv_time_ind = pd.DataFrame()
for loc in riv_dates.columns.tolist():
    riv_dates[loc] = pd.to_datetime(riv_dates[loc], format = "%m/%d/%Y") #convert to dt object
    loc_ind = []
    
    for event in riv_dates[loc]:
        ind = em.find_closest_val(event, par_df['time'])
        loc_ind.append(ind)
        
    riv_time_ind[loc] = loc_ind
```

```python
#find time indices to get 
time_get = []
import numpy as np
for val in np.asarray(riv_time_ind).flatten():
    if val not in time_get:
        for i in range(3):
            time_get.append(val - i)
time_get = np.sort(time_get)
```

```python
bbox = {'lat' : [16, 54],
        'lon' : [-100, -62]}

for i in range(len(bbox['lon'])):
    val = bbox['lon'][i]
    if val < 0:
        bbox['lon'][i] = 180 - val
if bbox['lon'][0] > bbox['lon'][1]: bbox['lon'] =  [bbox['lon'][1], bbox['lon'][0]]

bbox_c = {'lat': [em.find_closest_val(i, par_df['lat']) for i in bbox['lat']],
         'lon': [em.find_closest_val(i, par_df['lon']) for i in bbox['lon']]}

bbox_c
```

#### Get all data

```python
import time as pytime

data = {}
time_acq = np.arange(0, max(time_get) + 1)
print(max(time_get))

for v in var_df:
    url = url_base + var_df[v]['location']
    x = open_url(url)

    start_time = pytime.time()

    array = np.zeros((len(time_acq), bbox_c['lat'][1] - bbox_c['lat'][0], bbox_c['lon'][1] - bbox_c['lon'][0]))
    
    for i in range(len(time_acq)):
        t = time_acq[i]


        if var_df[v]['level']:
            d = x[v][v][int(t), 0,  bbox_c['lat'][0]:bbox_c['lat'][1], bbox_c['lon'][0]:bbox_c['lon'][1]]
        else: 
            d = x[v][v][int(t), bbox_c['lat'][0]:bbox_c['lat'][1], bbox_c['lon'][0]:bbox_c['lon'][1]]

        d = np.squeeze(d)
        array[i, :, :] = d

    data[v] = array 

    print(v + ' acquired: ' + str(pytime.time() - start_time) )
```

#### Formatting data

```python
# remove nans
for v in data:
    data[v][data[v] < -10**30 ] = np.nan
```
